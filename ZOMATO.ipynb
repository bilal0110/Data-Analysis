### Reading the Country Code file


```python
df2=pd.read_excel('Zomato_Country-Code.xlsx')
df2
```

### Merging the country code file into zomato data base
 

```python
db=pd.merge(df,df2,on='Country Code',how='left')
db
```

## Rating Aggregation


```python
db['Rating text'].value_counts()
```

```python
db.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings=db.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Ratings Count'})
```

```python
ratings
```

```python
import matplotlib as m
m.rcParams['figure.figsize']=(12,8);
sns.barplot(ratings['Aggregate rating'],ratings['Ratings Count'],data=ratings);
```

```python
m.rcParams['figure.figsize']=(12,8)
sns.barplot(x="Aggregate rating",y="Ratings Count",hue='Rating color',data=ratings,palette=['black','red','orange','yellow','green','green'])
```

## Frequency Distribution of rating color in data


```python
#Frequency Distribution of rating color in data
# Count plot
sns.countplot(ratings['Rating color'],data = ratings, palette = ['black','red','orange','yellow','green','green']);
# plt.xticks(['0.0'.''])
```
